```python
# STEP 1: Install dependencies
!pip install transformers accelerate -q

# STEP 2: Import the model pipeline
from transformers import pipeline

# STEP 3: Load Starcoder2 model
code_gen = pipeline("text-generation", model="bigcode/starcoder2-3b")

# STEP 4: Define a function to generate code
def generate_code_response(task):
    prompt = f"# Python task: {task}\n\n"
    result = code_gen(prompt, max_new_tokens=150, temperature=0.7, do_sample=True)[0]['generated_text']
    return result.replace(prompt, "").strip()

# STEP 5: Add your questions to this list
tasks = [
    # Your 10 Tasks
    "Check if a number is positive.",
    "Check whether a number is even or odd.",
    "Find the larger of two numbers.",
    "Assign a grade based on marks: Marks â‰¥ 90: Grade A, â‰¥ 75 and < 90: Grade B, â‰¥ 50 and < 75: Grade C, < 50: Grade F.",
    "Check if a person is eligible to vote (age â‰¥ 18).",
    "Check if a number is negative, positive, or zero.",
    "Check if a given year is a leap year.",
    "Determine whether a person is a child, teenager, adult, or senior based on age.",
    "Check if a character entered is a vowel or consonant.",
    "Compare three numbers and find the largest one.",

    # Previously listed tasks
    "Write a Python function to calculate the factorial of a number",
    "Write a Python script to reverse a string",
    "Create a list of prime numbers up to 100",
    "Sort a list in ascending order",
    "Write a function to check if a string is a palindrome",
    "Create a dictionary from two lists",
    "Print the Fibonacci sequence up to n terms",
    "Count vowels in a given string",
    "Find the maximum element in a list",
]

# STEP 6: Run code generation for each task
for i, task in enumerate(tasks, start=1):
    print(f"\nðŸ§  Task {i}: {task}")
    print("ðŸ“œ Generated Code:\n")
    print(generate_code_response(task))
    print("=" * 60)
```

```python
import gradio as gr
interface = gr.Interface(
    fn=generate_code_response,
    inputs=gr.Textbox(lines=2, placeholder="Describe your Python task..."),
    outputs=gr.Code(language="python"),
    title="ðŸ¤– AI Coding Assistant",
    description="Enter a simple coding problem in English. The AI will generate Python code using Starcoder2."
)

interface.launch(share=True)  # Use share=True to get a public link
```
